```python
import pandas as pd
import plotly.express as px
```

#### Importando as bases de dados

```python
df_principal = pd.read_excel('acoes_pura.xlsx', sheet_name = 'Principal')
df_principal
```

```python
df_total_acoes = pd.read_excel('acoes_pura.xlsx', sheet_name = 'Total_de_acoes')
df_total_acoes
```

```python
df_ticker = pd.read_excel('acoes_pura.xlsx', sheet_name = 'Ticker')
df_ticker
```

```python
df_chatgpt = pd.read_excel('acoes_pura.xlsx', sheet_name = 'Chatgpt')
df_chatgpt
```

#### Agregando informações a base de dados principal

```python
df_principal.columns
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns = {'Último (R$)': 'Valor_final', 'Var. Dia (%)': 'Var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal.Var_dia_pct / 100
df_principal
```

```python
df_principal['Valor_inicial'] = df_principal.Valor_final / (1 + df_principal.Var_pct)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on = 'Ativo', right_on = 'Código', how = 'left')
df_principal
```

```python
df_principal = df_principal.drop(columns = ['Código'])
df_principal
```

```python
df_principal = df_principal.rename(columns = {'Qtde. Teórica': 'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal.Valor_final - df_principal.Valor_inicial) * df_principal.Qtd_teorica
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal
```

```python
df_principal.Qtd_teorica = df_principal.Qtd_teorica.astype(int)
```

```python
df_principal
```

```python
df_principal['Resultado'] = df_principal.Variacao_rs.apply(lambda x: 'Subiu' if x > 0 else 'Desceu' if x < 0 else 'Estável')
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on = 'Ativo', right_on = 'Ticker', how = 'left')
df_principal = df_principal.drop(columns = ['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on = 'Nome', right_on = 'Empresa', how = 'left')
df_principal = df_principal.drop(columns = ['Empresa'])
df_principal = df_principal.drop_duplicates()
df_principal
```

```python
df_principal = df_principal.rename(columns = {'Idade (anos)': 'Idade'}).copy()
df_principal
```

```python
df_principal['Cat_idade'] = df_principal.Idade.apply(lambda x: 'Mais de 100 anos' if x > 100 else 'Menos de 50 anos' if x < 50 else 'Entre 50 e 100 anos')
df_principal
```

#### Exibindo informações gerais

```python
maior = df_principal.Variacao_rs.max()
menor = df_principal.Variacao_rs.min()
media = df_principal.Variacao_rs.mean()
media_subiu = df_principal[df_principal.Resultado == 'Subiu']['Variacao_rs'].mean()
media_desceu = df_principal[df_principal.Resultado == 'Desceu']['Variacao_rs'].mean()

print(f'Maior: R$ {maior:,.2f}')
print(f'Menor: R$ {menor:,.2f}')
print(f'Média: R$ {media:,.2f}')
print(f'Média (+): R$ {media_subiu:,.2f}')
print(f'Média (-): R$ {media_desceu:,.2f}')
```

#### Analisando e criando gráficos para grupos de dados

```python
df_principal_subiu = df_principal[df_principal.Resultado == 'Subiu']
df_principal_subiu.head()
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
fig = px.pie(df_analise_segmento,
             values = 'Variacao_rs',
             names = 'Segmento',
             title = 'Variação em Reais (R$) por Segmento',
             width = 800,
             height = 600)
fig.show()
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo,
             x = 'Resultado',
             y = 'Variacao_rs',
             text = 'Variacao_rs',
             title = 'Variação em Reais (R$) por Resultado')
fig.show()
```

```python
df_analise_idade = df_principal.groupby('Cat_idade')['Variacao_rs'].sum().reset_index()
df_analise_idade
```

```python
df_qte_empresas = df_principal.groupby('Cat_idade').size().reset_index()
df_qte_empresas = df_qte_empresas.rename(columns = {0 : 'Qtd_empresas'}).copy()
df_qte_empresas
```

```python
df_analise_idade = pd.concat([df_analise_idade, df_qte_empresas.Qtd_empresas], axis = 1)
df_analise_idade
```

```python
fig = px.bar(df_analise_idade,
             x = 'Cat_idade',
             y = 'Variacao_rs',
             text = 'Variacao_rs',
             title = 'Variação em Reais (R$) por Categoria de idade',
             labels= {'Cat_idade': 'Categoria de idades',
                     'Variacao_rs': 'Variação em reais (R$)'})
fig.update_traces(textposition = 'outside')
fig.show()
```
